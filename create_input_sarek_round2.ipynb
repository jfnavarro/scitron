```python
import pandas as pd
import boto3
import os

#Â Parse Excel Exomes
meta = pd.read_excel("Metadata2.xlsx")
```

```python
meta[meta['ROUND'] == 'SECOND']
```

```python
samples_to_process = list()
for index, row in meta[meta['ROUND'] == 'SECOND'].iterrows():
    stype = row['SAMPLE_TYPE']
    patient = row['PATIENT_ID']
    sample = row['SAMPLE_ID']
    if stype != 'gDNA':
        samples = meta.loc[(meta['PATIENT_ID'] == patient)
                           & (meta['SAMPLE_TYPE'] == 'gDNA')
                           & (meta['ROUND'] == 'FIRST'), 'SAMPLE_ID']
        for s in samples:
            samples_to_process.append(s)
    else:
        samples = meta.loc[(meta['PATIENT_ID'] == patient)
                           & (meta['SAMPLE_TYPE'] != 'gDNA')
                           & (meta['ROUND'] == 'FIRST'), 'SAMPLE_ID']
        for s in samples:
            samples_to_process.append(s)
    samples_to_process.append(sample)
meta.loc[meta['SAMPLE_ID'].isin(samples_to_process), :]
```

```python
# Sarek format
sub_meta = meta.loc[meta['SAMPLE_ID'].isin(samples_to_process),
                    ["PATIENT_ID", "GENDER", "SAMPLE_TYPE", "SAMPLE_ID", "LANE", "FASTA1", "FASTA2"]]

#Â Convert sample type to 0 (normal) and 1 (tumor)
sub_meta.loc[sub_meta["SAMPLE_TYPE"] == "gDNA", "SAMPLE_TYPE"] = 0
sub_meta.loc[sub_meta["SAMPLE_TYPE"] != 0, "SAMPLE_TYPE"] = 1

filenames = set()
files1 = set()
files2 = set()
s3 = boto3.client('s3')
for obj in s3.list_objects_v2(Bucket='scitron', Prefix='exomes/')['Contents'][1:]:
    filenames.add(os.path.basename(obj['Key']))
for index, row in sub_meta.iterrows():
    file1 = os.path.basename(str(row['FASTA1']))
    files1.add(file1)
    file2 = os.path.basename(str(row['FASTA2']))
    files2.add(file2)
    if file1 not in filenames or file2 not in filenames:
        print("Row {} does not have the correct file names {} {}".format(index, file1, file2))
```

```python
#Â Write metadata
sub_meta.to_csv("samples2.tsv", sep="\t", header=None, index=None)
```
